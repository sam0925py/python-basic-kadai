```python

# 消費税を加えた金額を計算する関数
def calculate_total_price(price: int, tax_rate: int) -> int:
    tax_amount = price * tax_rate / 100
    total = price + tax_amount
    return int(total)  # 小数点以下を切り捨てて整数で返す

# 関数の実行例
product_price = 110      # 商品の金額
consumption_tax = 10     # 消費税率（%）

# 関数を使って合計金額を計算
total_price = calculate_total_price(product_price, consumption_tax)

# 結果を表示
print(f"税込価格は {total_price} 円です。")
```
