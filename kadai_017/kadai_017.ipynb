名前(name)と年齢(age)の属性を持つHumanクラスを作成してください。

Humanクラスには、以下の条件で標準出力(print)するcheck_adultメソッドを追加してください。

ageが20以上の場合に大人であること
そうでない場合に大人でないこと
Humanクラスのインスタンスを複数生成してリストに追加し、リストの要素数分だけcheck_adultメソッドを呼び出してください

```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}（{self.age}歳）は大人です。")
        else:
            print(f"{self.name}（{self.age}歳）は大人ではありません。")

# Humanクラスのインスタンスを複数生成
humans = [
    Human("太郎", 25),
    Human("花子", 18),
    Human("次郎", 20),
    Human("三郎", 15)
]

# リストの要素数分だけcheck_adultメソッドを呼び出す
for human in humans:
    human.check_adult()
```
